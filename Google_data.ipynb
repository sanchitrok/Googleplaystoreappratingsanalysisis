# Data Imputation & Manipulation

### Fill the null values with appropriate values using aggregate functions such as median, mode or mean

```python
#define a function impute_median
def impute_median(series):
    return series.fillna(series.median())
```

```python
google_data.Rating=google_data['Rating'].transform(impute_median)
```

```python
google_data.isnull().sum()        #count the number of null values in each column
```

```python
# check modes of categorical columns
print(google_data['Type'].mode())
print(google_data['Current Ver'].mode())
print(google_data['Android Ver'].mode())
```

```python
google_data['Type'].fillna(str(google_data['Type'].mode().values[0]),inplace=True)
google_data['Current Ver'].fillna(str(google_data['Current Ver'].mode().values[0]),inplace=True)
google_data['Android Ver'].fillna(str(google_data['Android Ver'].mode().values[0]),inplace=True)
```

```python
google_data.isnull().sum()
```

```python
# convert Ratings, Reviews, Price into numerical values
google_data['Price']=google_data['Price'].apply(lambda x:str(x).replace('$','') if '$' in str(x) else str(x))
google_data['Price']=google_data['Price'].apply(lambda x:float(x))
```

```python
google_data['Reviews']=pd.to_numeric(google_data['Reviews'],errors='coerce')
```

```python
google_data['Rating']=google_data['Rating'].apply(lambda x:str(x).replace('+','') if '+' in str(x) else str(x))
google_data['Rating']=google_data['Rating'].apply(lambda x:str(x).replace(',','') if ',' in str(x) else str(x))
google_data['Rating']=google_data['Rating'].apply(lambda x:float(x))
```

#### Data Visualization

```python
grp=google_data.groupby('Category')
x=grp['Rating'].agg(np.mean)
y=grp['Price'].agg(np.sum)
z=grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,'ro')
plt.xticks(rotation=90)
plt.title('Category wise Ratings')
plt.xlabel('Categories')
plt.ylabel('Ratings')
plt.show()
```

```python
#Observations: Categories like Education, Events & Art&Design enjoys a higher ratings compared to other categories
```

```python
plt.figure(figsize=(12,6))
plt.plot(y,'bs')
plt.xticks(rotation=90)
plt.title('Category wise Pricing')
plt.xlabel('Categories')
plt.ylabel('Price in $')
plt.show()
```

```python
#Observations: Categories like Finance, Family & Lifestyle have a higher pricing compared to other categories
```

```python
plt.figure(figsize=(12,6))
plt.plot(z,'g^')
plt.xticks(rotation=90)
plt.title('Category wise Reviews')
plt.xlabel('Categories')
plt.ylabel('Reviews')
plt.show()
```

```python
#Observations: Categories like Communication & Social have higher reviews compared to other categories
```
